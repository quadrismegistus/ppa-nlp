## Getting the counts using db

```python
def get_counts_from_db(self):
    return {
        wwc.word:wwc.count
        for wwc in WordCount.objects.filter(work=self)
    }

DigitizedWork.get_counts_from_db = get_counts_from_db
```

```python
work.get_counts_from_db()
```

```python
def get_counts(self,save=True):
    countd = self.get_counts_from_db()
    if not countd:
        countd = self.calc_counts()
        if save: self.store_counts(countd=countd)
    
    # reordered by count, in Counter container
    nice_counts = Counter(dict(Counter(countd).most_common()))
    return nice_counts

DigitizedWork.get_counts = get_counts
```

```python
work.get_counts()
```

```python
def get_total_count(self):
    if not hasattr(self,'_totalcount') or self._totalcount is None:        
        try:
            self._totalcount=self.totalcount_set.first().count
        except Exception as e:
            logging.debug(e)
            import numpy as np
            self._totalcount = np.nan
    return self._totalcount

DigitizedWork.get_total_count = get_total_count
```
